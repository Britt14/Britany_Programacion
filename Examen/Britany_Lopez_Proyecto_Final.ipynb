Contiene datos de emisiones de di√≥xido de carbono (CO‚ÇÇ) de varios pa√≠ses entre 1975 y 2017.

üßæ Estructura de los datos
- Country: Pa√≠s

- Type: Tipo de emisi√≥n:

- Per c√°pita: CO‚ÇÇ promedio por persona (t CO‚ÇÇ/persona)

    - Se calcula dividiendo emisiones totales por la poblaci√≥n

```python
#Importa el CSV como un DataFrame
import pandas as pd
df = pd.read_csv('Estimaciones_Emisiones_Dioxido_Carbono.csv', encoding='UTF=8')
df.head()
```

# Pregunta 0

1. Crear un dataframe con el nombre de variable 'df_emisions_co2' usando √∫nicamente las filas cuya columna 'Type' tiene el valor 'Emisiones totales de CO‚ÇÇ emitidas por el pa√≠s'

```python
#Crea un copia del original filtrado como df_emisions_co2
df_emisions_co2 = df[
    df['Type'] == 'Emisiones (kilotoneladas metricas de dioxido de carbono)'
]
df_emisions_co2.head()
```

```python
#Crea un copia del original filtrado como df_emisions_co2_per_capita
df_emisions_co2_per_capita = df[
    df['Type'] == 'Emisiones per capita (toneladas metricas de dioxido de carbono)'
]
df_emisions_co2_per_capita.head()
```

# Pregunta 9

Cu√°ntas son las emisiones totales en 'libras' s√≥lo para Costa Rica usando el dataframe 'df_emisions_co2'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
Total_Emisiones_Costa_Rica = df_emisions_co2[df_emisions_co2['Country'] == 'Costa Rica']['Value'].sum()

# Conversi√≥n de kilotoneladas m√©tricas a libras
Total_Emisiones_Libras = Total_Emisiones_Costa_Rica * 2_204_620  
print(f"La cantidad total de emisiones registradas en Costa Rica en libras es: {round(Total_Emisiones_Libras, 2)}.")
```

# Pregunta 10

Para s√≥lo 'Costa Rica' y suponiendo que las emisiones se estancan a partir del √∫ltimo a√±o reportado (se mantiene igual). Cu√°ntas son las emisiones en 'Kg' s√≥lo para Costa Rica desde el ultimo a√±o registrado hasta el 2025 usando el dataframe 'df_emisions_co2'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

#Se obtienen los datos de Costa Rica para evitar filtrar varias veces
df_Costa_Rica = df_emisions_co2[df_emisions_co2['Country'] == 'Costa Rica']

Emisiones_Ultimo_Anio = df_Costa_Rica[
    df_Costa_Rica['Year'] == df_Costa_Rica['Year'].max()
    ]['Value'].iloc[0]

#Determinar el √∫ltimo a√±o registrado
Ultimo_Anio = df_emisions_co2['Year'].max()

#Determinar los a√±os transcurridos desde el √∫ltimo a√±o registrado hasta el a√±o actual 
Ultimo_Anio_Hasta_Actual = 2025 - Ultimo_Anio + 1

#Convertir la cantidad de emisiones a kilogramos
Emisiones_Kilogramos = Emisiones_Ultimo_Anio * 1_000_000 * Ultimo_Anio_Hasta_Actual

print(f"La cantidad de emisiones registradas desde el √∫ltimo a√±o hasta el a√±o actual en Costa Rica es: {Emisiones_Kilogramos}.")
```

# Pregunta 12

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones totales en Centroam√©rica usando el dataframe 'df_emisions_co2'?
Se espera un dataframe de 2 columnas: 'Year' y 'total_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

Paises_Centroamericanos = ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua']
Emisiones_Centroamerica = df_emisions_co2[
    df_emisions_co2['Country'].isin(Paises_Centroamericanos)
]
Total_Emisiones_Centroamerica = Emisiones_Centroamerica.groupby('Year')['Value'].sum().reset_index(name= 'Total por a√±o')
print("Total de emisiones por a√±o en los pa√≠ses de Centroam√©rica:")
print(Total_Emisiones_Centroamerica)
```

# Pregunta 13

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones en promedio en Centroam√©rica usando el dataframe 'df_emisions_co2_per_capita'?
Se espera un dataframe de 2 columnas: 'Year' y 'promedio_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
Emisiones_Centroamerica_Per_Capita = df_emisions_co2_per_capita[
    df_emisions_co2_per_capita['Country'].isin(Paises_Centroamericanos)
]
Total_Emisiones_Centroamerica_Per_Capita = Emisiones_Centroamerica_Per_Capita.groupby('Year')['Value'].mean().reset_index(name= 'Promedio por a√±o')
print("Promedio de emisiones per capita  registradas por a√±o en los pa√≠ses de Centroam√©rica:")
print(Total_Emisiones_Centroamerica_Per_Capita)
```

# Pregunta 14

Cu√°l es el promedio de las emisiones per capita para el a√±o 2005 usando el dataframe 'df_emisions_co2_per_capita'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
Promedio_Emisiones_Per_Capita_2005 = df_emisions_co2_per_capita[
    df_emisions_co2_per_capita['Year'] == 2005
]['Value'].mean()
print(f"El promedio de emisiones per capita en el a√±o 2005 es: {Promedio_Emisiones_Per_Capita_2005}.")    
```

Usa el dataframe  'df_emisions_co2_per_capita' con las columnas Year y promedio_por_a√±o para crear una gr√°fica de l√≠neas que muestre la evoluci√≥n del promedio de emisiones per c√°pita en Centroam√©rica a lo largo del tiempo; el eje X debe ser el a√±o y el eje Y el promedio de emisiones.
